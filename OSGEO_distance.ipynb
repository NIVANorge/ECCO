## How change from lat/lon <--> Pixels[n,i]

```python
# Functions to transform between pixels and latitude points (and back again) based on coordinate system of GeoTiff
# Used http://nbviewer.ipython.org/github/profLewis/geogg122/blob/master/Chapter4_GDAL/GDAL_Python_bindings.ipynb
# unravel GDAL transform parameters
xoff, a, b, yoff, d, e = ds.GetGeoTransform()
print(xoff,a,b,yoff,d,e)

def pixel2coord(x, y):
    """Returns global coordinates from pixel x, y coords"""
    xp = a * x + b * y + xoff
    yp = d * x + e * y + yoff
    return(xp, yp)

def coord2pixel(lon,lat):
    """Return the pixels from the lon,lat coords"""
    pixel_x = int((lon - xoff)/a)
    pixel_y = int((lat - yoff)/(e))
    return(pixel_x,pixel_y)
```

## Gather Lake data and Find distance (with heights)

```python
meta = pd.read_csv('Metadata/Meta_Lakes.csv')    # Very nice!
meta.index = meta.hex                            # Index the metadata by lake ID's
lake_file = '/Users/Ben/Work/Python/Programs/ECCO/Lakes/Lake_v0.2/ecco-biwa_lakes_v.0.2.shp'
ShapeData = ogr.Open(lake_file)                  # Make a link to Lake Shape Files
TheLayer = ShapeData.GetLayer(iLayer=0)
```

```python
# Get only certain lakes
lon_means=[]
lat_means=[]
ID_list=[]
Alt_list=[]
Ctry_list=[]
Area_list=[]
#dolakes = range(0,275265,1)
dolakes = range(100)
for n in dolakes:
    feature1 = TheLayer.GetFeature(n)                               # Get individual lake in shapefile
    lake_feature = feature1.ExportToJson(as_object=True)
    if meta.area[lake_feature['properties']['EBhex'][2:]] > 0.001:  # If lakes are bigger than 2km^2 then get the shape
    #if lake_feature['properties']['polycountr'] == 'Sweden':       # Or if lakes are from certain country only
        print(n,end="\r")
        Alt_list.append(lake_feature['properties']['stf_mean'])
        ID_list.append(lake_feature['properties']['EBhex'][2:])
        Area_list.append(meta.area[lake_feature['properties']['EBhex'][2:]])
        Ctry_list.append(lake_feature['properties']['polycountr'])
        xpoints =[]                                                             # Tmp hold
        ypoints =[]
        for n,points in enumerate(lake_feature['geometry']['coordinates'][0]):
            xpoints.append(points[0])
            ypoints.append(points[1])
        lon_means.append(np.average(xpoints))
        lat_means.append(np.average(ypoints))        # For bokeh output needs structure: xs =[[lk1],[lk2]]
                                     # Note this element-wise structure even on metadata lists above  
    #ECCO.Preview_Lake(lake_cart)  
```

```python
print(lat_means[0],lon_means[0],Alt_list[0],ID_list[0])
```

```python
print('lon: %3.2f lat: %3.2f'%(lon_means[0],lat_means[0]))
xpix,ypix = coord2pixel(lon_means[0],lat_means[0])
print(xpix,ypix)
print(pixel2coord(xpix,ypix))
print('Distance (km)=',distance[ypix,xpix])
```

units,  kilometers from ocean coast, with uncertainty of 1 km. 
> 0 is ocean, < 0 is land 
0 is the coastline. 
Coast defined as GMT mapping software intermediate coastline
computation performed in GMT at 0.04 degrees
Bilinear interpolation to 0.01 degrees

```python
tmp = np.zeros([len(lon_means),5])

for n,i in enumerate(lon_means):
    tmp[n,0]=lon_means[n]
    tmp[n,1]=lat_means[n]
    tmp[n,2]=Alt_list[n]
    xpix,ypix = coord2pixel(lon_means[n],lat_means[n])   # This function finds the x/y pixels to pull out of data
    tmp[n,3]=(int(distance[ypix,xpix]))                  # here selecting the right pixel. 
    tmp[n,4]=Area_list[n]
```

```python
df = pd.DataFrame(tmp,index=ID_list,columns=['Lon','Lat','Alt','Dist','Area']) #Make a pd dataframe

df.to_csv('Metadata/Lake_Stats.csv',index_label='Lake_ID')  # Save output.
```
