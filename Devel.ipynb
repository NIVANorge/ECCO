## Read in GeoJSON data

```python
fn = 'Lakes/comsat_fetch.geojson'
driver = ogr.GetDriverByName('GeoJSON')
ds = driver.Open(fn)
lyr = ds.GetLayer('OGRGeoJSON')
dfn = lyr.GetLayerDefn()
fieldnames = [dfn.GetFieldDefn(i).GetName() for i in range(dfn.GetFieldCount())]

tst = []
for i in range(dfn.GetFieldCount()):
    tst.append(dfn.GetFieldDefn(i).GetName())
    print i, dfn.GetFieldDefn(i)
```

```python
import json   # use json module as same solution in 
# http://gis.stackexchange.com/questions/73768/how-to-convert-geojson-to-python-objects

# Also of use, this link shows how to access the http://geographika.co.uk/visualising-geojson-in-15-minutes
# data remotley with url2lib

with open('Lakes/comsat_fetch.geojson') as f:
    data = json.load(f)

lake_id =[]
lake_path=[]
lake_geometry=[]
lake_name=[]
for feature in data['features']:
    #print feature['geometry']['type']
    #print feature['id']
    #print feature['geometry']['coordinates']
    lake_geometry.append(feature['geometry']['type'])
    lake_id.append(feature['id'])
    lake_name.append(feature['properties']['comsat.Lake'])
    lake_path.append(feature['geometry']['coordinates'])

print 'Read data for', len(lake_id), ' lakes.'    

#Note how, the JSON files are actually a bunch of dictionary objects. Each lake (which itself 
# is a collection of various dictionaries), is called a feature. The features then have
# dictionaries defining properties, and 'geometry'.
```
