## Create Lake Polygons from read JSON data

To plot a polygon, it is necessary to have two lists, one of verts (x,y coordinates), and the other of 
codes (commands to begin, carry on, and close the polygon)

Note, the Path objects (Matplotlib.Path.Path), which are used here as the lake polygon objects are made up of two NP arrays.
The main components of which are Path.verticies[n], which list the x,y coordinates, and then the Path.codes[n], which describe
the relationship of the x,y coordinates (ie. start of a polygon, path, link, or end), so the object can be drawn.

You can therefore query, and call the data within the polygon using the Path.verticies array.


```python
# -- LAKE POLYGON CREATION --
# Input  - Lake coordinate data as the x,y list from GeoJSON.
# Output - a Matplotlib Path object, and also the lower/upper corners (with small offset)
#         of the lake lon/lats
# Requirements - Matplotlib.Path import Path
# Optional - plot the path with patches function (import Matplotlib.patches as patches)
#
# Note. The way the lake addressing goes is Lake[0][0][0] where the first 0 is the 
# individual lake. Lake[0][0][:] will give you every x,y element. Lake[0][0][0][0] will
# give the first x-element of the first lake. Lake[0][0][0][1] will give the first y-element etc.
#
def pth_create(coord):
    verts =[]
    codes=[]
    xonly =[]
    yonly =[]
    pth_ln = int(len(coord) - 1)
    #print shape(coord)
    #print verts,codes,pth_ln    
    for i,n in enumerate(coord[0:pth_ln]):
        #print i,n
        verts.append(n)
        if i == 0:
            codes.append(Path.MOVETO)
        if ((i >= 1)&(i < (pth_ln - 1))):
            codes.append(Path.LINETO)  
        if i == (pth_ln - 1):
            codes.append(Path.CLOSEPOLY)
    path_out = Path(verts, codes)
    # Now find the lat/lon edges
    for i in xrange(len(path_out.vertices)):
        xonly.append(path_out.vertices[i][0])
        yonly.append(path_out.vertices[i][1])   
    return path_out,min(xonly),max(xonly),min(yonly),max(yonly)
```

## Find physical area of Lakes (in $km^{2}$)

Asssume a spherical Earth is sufficent to use to calculate the lake areas in $km^{2}$.
Solution from stackoverflow.com/questions/4681737/how-to-calculate-the-area-of-a-polygon-on-the-earths-surface-using-python
Multiply the latitude by the length of one degree of latitude, and the longitude by the length of
a degree of latitude and the cosine of the latitude. Then, calculate the area of an arbitrary polygon in a plane.

```python
# Input the Matplolib.Path.Path object verticies
# Oputput x,y coridinates as if they have been projected onto a Sphere
#
from math import pi, cos, radians
def eqarea(verts):
    earth_radius = 6367.4447        # Earth avg. radius (km), Wolfram Alpha
    lat_dist = pi * earth_radius / 180.0  
    eqout =[]
    for n,i in enumerate(verts):
        longitude = i[0] 
        latitude = i[1]
        x = longitude * lat_dist * cos(radians(latitude))
        y = latitude * lat_dist
        #if n == 0:
        #    print ' Test view '
        #    print x,'=',longitude,'x',lat_dist,'x',cos(radians(latitude))
        #    print y,'=',latitude,'x',lat_dist
        #    print ' '
        eqout.append([x,y])
    return eqout
#
# This function implements Green's Theorem (http://code.activestate.com/recipes/578275-2d-polygon-area)
# to find the area of a polygon in a 2D co-ordinate system.
def poly_area2D(poly):
    total = 0.0
    N = len(poly)
    for i in range(N):
        v1 = poly[i]
        v2 = poly[(i+1) % N]
        total += v1[0]*v2[1] - v1[1]*v2[0]
    return abs(total/2.)   
```

```python
# Quick function to find the bounding lat/lons of each lake case:
# Input - verticies of lake border
# Output - x[max,min],y[max,min]
#
def get_latlonlim(xypath):
    xset = []
    yset = []
    for i in xypath:
        yset.append(i[1])
        xset.append(i[0])
    #print 'Returning Lake lon and lat max/min vals'
    return [max(xset),min(xset)],[max(yset),min(yset)]
```
